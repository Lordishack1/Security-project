```python
def is_ip(address):
    return address.replace('.', '').isnumeric()
hostname_list = []
ip_address_list = []
domain_list = []
for url in df['Indicator']:
    hostname_with_path = url.split("//")[1] # Task 1
    hostname_only = hostname_with_path.split("/")[0]
    if is_ip(hostname_only):
        ip_address_list.append(hostname_only)
    else:
        hostname_list.append(hostname_only)
        domain = tldextract.extract(hostname_only).domain
        domain_list.append(domain)

data_df_domain= pd.DataFrame(hostname_list, columns=['hostname'])
data_df_domain['domain'] = domain_list
len(ip_address_list) # Task 2
len(pd.unique(data_df_domain['domain'])) # Task 3
#data_df_domain['domain'].value_counts()
```

```python
hostname_list= []
tld_list=[]
subdomain_list=[]
domain_len_list=[]
subdomain_level_list=[]
vowels_count_list=[]
digit_count_list=[]
domain_subdomain = []

for url in df['Indicator']:
    hostname_with_path = url.split("//")[1] 
    
    hostname_only = hostname_with_path.split("/")[0]
    if not is_ip(hostname_only):
        hostname_list.append(hostname_only)
        tld = get_tld(hostname_only, fix_protocol=True)
        tld_list.append(tld)
        subdomain = tldextract.extract(hostname_only).subdomain
        subdomain_list.append(subdomain)
        domain = tldextract.extract(hostname_only).domain
        domain_len_list.append(len(domain))
        if subdomain == '':
            subdomain_level_list.append(0)
        else:
            subdomain_level_list.append(len(subdomain.split(".")))
        vowels_count_domain = [domain.count(x) for x in "aeiou"]
        vowels_count_list.append(sum(vowels_count_domain))
        digits_count_domain = [domain.count(x) for x in "1234567890"]
        digit_count_list.append(sum(digits_count_domain))
        if subdomain == '':
            domain_subdomain.append(domain)
        else:
            domain_subdomain.append(domain + "." + subdomain)
        
data_df= pd.DataFrame(hostname_list, columns=['hostname'])
data_df['TLD']= tld_list
data_df['subdomain'] = subdomain_list
data_df['domain_len'] = domain_len_list
data_df['subdomain_level'] = subdomain_level_list
data_df['vowel_counts']= vowels_count_list
data_df['digit_counts']= digit_count_list
data_df['keywords'] = domain_subdomain
```
