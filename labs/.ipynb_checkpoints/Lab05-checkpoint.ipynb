```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import reverse_geocode
from geopy.geocoders import Nominatim
zeroaccess_df = pd.read_csv("zeroaccess.csv")
state_internet_df = pd.read_csv("state-internets.csv")
county_data_df = pd.read_csv("county-data.csv")
```

```python
zeroaccess_df["lat"] = zeroaccess_df["lat"].astype(str)
zeroaccess_df["long"] = zeroaccess_df["long"].astype(str)

zeroaccess_df["location"] = zeroaccess_df["lat"] + ", " + zeroaccess_df["long"]
```

```python
country_list= []
for (x,y) in zip(zeroaccess_df['lat'], zeroaccess_df['long']): # iterating over 2 columns
    lat= float(x) 
    long=float(y)
    country_list.append(reverse_geocode.get((lat, long))['country'])
len(country_list)    
zeroaccess_df['country'] = country_list
```

```python
us_zeroaccess_df = zeroaccess_df[zeroaccess_df['country'] == 'United States']
us_zeroaccess_df
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
state_list = []
county_list = []
for (x,y) in zip(us_zeroaccess_df['long'][:10], us_zeroaccess_df['lat'][:10]):
    lat = str(y)
    long = str(x)
    state_list.append(geolocator.reverse(long+','+lat).raw['address']['state'])
    county_list.append(geolocator.reverse(long+","+lat).raw['address']['county'])
us_zeroaccess_df['state'] = state_list
us_zeroaccess_df['county'] = county_list
#Never figured out how to make this infinate loop stop. plus never had the correct synatx for it
```
