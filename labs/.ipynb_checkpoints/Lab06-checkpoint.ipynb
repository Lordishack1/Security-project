```python
login_list = []
office_home_list = []
home_list = []
bad_login_list = []
for (x,y) in zip(datalog_df["Longitude"],datalog_df["Latitude"]):
    loca = (str(x),str(y))
    login_list.append(loca)

for (x,y) in zip(office_locations_df["Longitude"],office_locations_df["Latitude"]):
    loca_office = (str(x),str(y))
    office_home_list.append(loca_office)

for (x,y) in zip(employee_details_df["Home Longitude"],employee_details_df["Home Latitude"]):
    loca_homes = (str(x),str(y))
    home_list.append(loca_homes)

for (x,y,z) in zip(login_list,office_home_list,home_list):
    if geodesic(x,y).miles > 10:
        bad_login_list.append(x)
    elif geodesic(x,z).miles > 10:
        bad_login_list.append(x)
bad_login_list = list(set(bad_login_list))
bad_login_list
```

If an employee trys to get resources from other departments

```python
resource_dept_dict = {}
id_dept_dict = {}
bad_resource_list = []
bad_employee_list = []
bad_list = []
for (x,y) in zip(dept_resources_df["ResourceID"], dept_resources_df["Department"]):
    resource_dept_dict[x] = y 
    
for (x,y) in zip(employee_details_df["EmployeeID"],employee_details_df["Department"]):
    id_dept_dict[x] = y
    
for (x,y) in zip(datalog_df["EmployeeID"],datalog_df["ResourceID"]):
    if y != 'AuthServer':
        if id_dept_dict[x] != resource_dept_dict[y]:
            bad_resource_list.append(y)
            bad_employee_list.append(x)
for (x,y) in zip(bad_employee_list, bad_resource_list):
    bad_peeps_place = (x,y)
    bad_list.append(bad_peeps_place)
bad_list
```
