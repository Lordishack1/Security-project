```python
import pandas as pd
import ipaddress
import matplotlib.pyplot as plt
import re
import seaborn as sns
ip_reputation_df = pd.read_csv("AlienVault_IP_reputation.csv")
zbot_df = pd.read_csv("zeus-book.csv", skiprows=4)
ip_space_df = pd.read_csv("ipv4-address-space.csv")
```

```python
zbot_df = zbot_df.rename(columns = {'##############################################################################':'IP'})
```

```python
ip_reputation_df["subnet_8"] = ip_reputation_df["IP"].str.lstrip('0')
ip_reputation_df["subnet_8"] = ip_reputation_df["subnet_8"].str.split(".").str[:1].str.join(".")
ip_reputation_df["subnet_8"] = ["0.0.0.0/8" if x=="" else x + ".0.0.0/8" for x in ip_reputation_df["subnet_8"]]
ip_reputation_df
```

```python
ip_space_df["subnet_8"] = ip_space_df["Prefix"].str.lstrip('0')
ip_space_df["subnet_8"] = ip_space_df["subnet_8"].str.split("/").str[:1].str.join(".")
ip_space_df["subnet_8"] = ["0.0.0.0/8" if x=="" else x + ".0.0.0/8" for x in ip_space_df["subnet_8"]]
ip_space_df
```

```python
new_merged_df = pd.merge(ip_reputation_df, ip_space_df, on=['subnet_8'])
new_merged_df
```

```python
ip = []
count = 0
ip = zbot_df['IP'].to_list()
for x in ip_reputation_df['IP']:
    if x in ip:
        count += 1
print(count)
```

```python
zbot_df["subnet_8"] = zbot_df["IP"].str.lstrip('0')
zbot_df["subnet_8"] = zbot_df["subnet_8"].str.split(".").str[:1].str.join(".")
zbot_df["subnet_8"] = ["0.0.0.0/8" if x=="" else x + ".0.0.0/8" for x in zbot_df["subnet_8"]]
merged = pd.merge(new_merged_df, zbot_df, on=['subnet_8'])
merged
```

```python
new_merged_ip_df = pd.merge(zbot_df,new_merged_df, on=['IP'])
new_merged_ip_df
```

```python
new_merged_ip_df['Country'].value_counts().plot(kind='bar');
```

```python
new_merged_ip_us_df = merged[merged['Country'].isin(['us'])] 

sns.lmplot(x='Risk', y= 'Reliability', data= new_merged_ip_us_df)
```
